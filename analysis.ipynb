```python
from pathlib import Path

import pandas as pd

DATASETS = ["arxiv_2023", "ogbn-arxiv", "cora", "pubmed"]
SEEDS = range(4)
MODELS = ["MLP", "GCN", "SAGE", "RevGAT", "LM"]
FEATURES = ["TA", "P", "E", "None"]


RESULT_DIR = Path(f"sep_cas_rets")

def get_result_path(dataset: str, seed: int, model: str, feature: str, split: str) -> Path:
    return RESULT_DIR / f"{dataset}/{seed}/{model}" / f"{feature}_raw_change_{split}.csv"
```

```python
result_df = pd.DataFrame()
for dataset in DATASETS:
    for model in MODELS:
        for feature in FEATURES:
            df = pd.DataFrame()
            for seed in SEEDS:
                result_path = get_result_path(dataset, seed, model, feature, "test")
                new_df = pd.read_csv(result_path)
                new_df = new_df.sum(axis=0).to_frame().T
                df = pd.concat([df, new_df], axis=0)
            df = df.mean(axis=0)
            df["dataset"] = dataset
            df["model"] = model
            df["feature"] = feature
            result_df = pd.concat([result_df, df.to_frame().T], axis=0)
result_df = result_df.set_index(["dataset", "model", "feature"])
result_df.to_csv("analysis.csv")
result_df
```
